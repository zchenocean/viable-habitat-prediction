## calculate the Metabolic Index for each grid point at each depth for FOSI hindcast

```python
import metabolic as mi
import xarray as xr
import numpy as np
import netCDF4
import os
import warnings
warnings.filterwarnings('ignore')
import matplotlib.colors as colors
from matplotlib import cm
import matplotlib.pyplot as plt
```

```python
USER = os.environ['USER']
path = f'/glade/scratch/{USER}/DPLE-results'
tres = 'monthly'
#tres = 'yearly'
```

```python
filename = f'{path}/DPLE-FOSI_hindcast_T_S_O2_1948-2017_{tres}.nc'
ds1 = xr.open_dataset(filename)
Temp = ds1['TEMP']
#
filename = f'{path}/DPLE-FOSI_hindcast_pO2_1948-2017_{tres}.nc'
ds2 = xr.open_dataset(filename)
pO2 = ds2['pO2']
pO2
```

## Calculate Φ using 13 different Eo values and same Ac value

```python
Ac_dist = [10]
Eo_dist = np.arange(-.2,1.1,.1)
print(Ac_dist)
print(Eo_dist)
```

```python
%%time
nrow = len(Ac_dist)
ncol = len(Eo_dist)
from itertools import product
dss = xr.Dataset()

for i, j in product(range(nrow), range(ncol)):
    Ac = Ac_dist[i]
    Eo = Eo_dist[j]
    print(Ac,Eo)
    dss[f'Phi_{nrow*i+j}'] = mi.Phi(pO2, Temp, Ac, Eo, dEodT=0.022)
dss
```

```python
dout = f'/glade/scratch/{USER}/DPLE-results'
os.makedirs(dout, exist_ok=True)
dss.load()
```

```python
%%time
dss.to_netcdf(f'{dout}/DPLE-FOSI_hindcast_Phi_Eo_space_1948-2017_{tres}.nc', mode='w')
```
